# how to delete a column??

```python
df.drop('cost_to_company')
```

```python
# KeyError: "['cost_to_company'] not found in axis"--> what does axis mean in previous error?
# by default 'Axis' = 0 it means that we are trying to delete a row named 'cost_to_company'. To delete a column we need to make 'Axis' = 1.
df.drop('cost_to_company', axis = 1)
```

```python
# here we have cost_to_company bcz once we perform drop() operation the column gets deleted from the current instances but not from the original data frame.
df.head()
```

```python
# what to do if we wanr to make permanent changes in the original data frame? There are 2 ways--> 1. by using assignment  2. using parameter inplace i.e inplace = True
df.drop('cost_to_company', axis = 1, inplace = True)
```
