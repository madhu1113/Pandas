# Difference between merge and join operation in pandas

```python
# merge- It will merge the common data. By default it performs inner join.
pd.merge(df4, df5)
```

```python
# it is sililer to left join.
pd.merge(df4, df5, how='left')
```

```python
# It is similar to right join operation.
pd.merge(df4, df5, how='right')
```

```python
# it is similar to right join operation.
pd.merge(df4, df5, how='outer')
```

```python
df4
```

```python
df5
```

```python
# it is similar to cross join operation
pd.merge(df4, df5, how='cross')
```

```python
# It performs right join on emp_id i.e common between both df4 and df5.
pd.merge(df4, df5, how='right', on= 'emp_id')
```

# Create a new column called profit_flag and it should be derived from profit column which says whether profit is "positive" and "negative"

```python
def profit_flag(a) : 
    if a >= 0 :
        return 'positive' 
    else :
        return 'negative'
```

```python

profit_flag(-45)
```

```python
profit_flag(45)
```

```python
# apply() is a function which gets applied on each element of the particular column. Inside apply we pass user defined function or lambda function or inbuilt function.
# It is similar to map(), reduce() function in list.
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

# create a new column which stores the length of customer_name.

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

# Create a new column which determines whether the the quantity is low, medium or high.

```python
def quantity_flag(a) : 
    if  a < 10 :
        return 'low'
    elif a >10 and a <20 :
        return 'medium'
    else :
        return 'high'
```

```python
df_sales['flag_quantity']= df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

# Create a new column and store the square of quantity inside that.

```python

df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a: a**2)
```
